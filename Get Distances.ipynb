```python
crime_records = []
for index, row in crimes.iterrows():
    if str(row['Latitude']) != "nan" and str(row['Longitude']) !="nan":
        crime_coords = (row['Latitude'], row['Longitude'])
        crime_records.append(crime_coords)
```

```python
def count_crimes(loc_coords, crime_locations, distMark):
    count = 0
    #print(loc_coords)
    for item in crime_locations:
        print(item)
        
        dist = geodesic(item, loc_coords)
        if dist <distMark:
            count+=1
            
    return count
    
```

```python
def getCrimesFromAddress(address, distance):
    geocoderApi = herepy.GeocoderApi('qXRCzE6fLQqNBe5NsF8f', 'tnWCtQU3V1fAWBpU9BiWeg')
    response = geocoderApi.free_form(address)
    response = json.loads(str(response))
    response = (response['Response']["View"][0]['Result'][0]['Location']['DisplayPosition'])
    response = (float(response['Latitude']), float(response['Longitude']))
    #print(response)
    print(count_crimes(response, crime_records, distance))
```
